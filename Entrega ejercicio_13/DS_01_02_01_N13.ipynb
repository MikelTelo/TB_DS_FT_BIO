## Ejercicio 2
Haz un programa que detecte todos los duplicados de un elemento en una lista:
* Utiliza una variable duplicado para asignarle el valor del elemento del que queremos encontrar todos sus duplicados.
* El programa debe devolver todos los índices de los elementos duplicados.
* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
def find_duplicates(lst, element):
    indices = []
    element_lower = element.lower()
    for index, value in enumerate(lst):
        if value.lower() == element_lower:
            indices.append(index)
    return indices

ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

# Buscar duplicados de "un"
duplicados_un = find_duplicates(ejer_3, "un")
print(f'Duplicados de "un": {duplicados_un}')

# Buscar duplicados de "es"
duplicados_es = find_duplicates(ejer_3, "es")
print(f'Duplicados de "es": {duplicados_es}')

# Buscar duplicados de "binario"
duplicados_binario = find_duplicates(ejer_3, "binario")
print(f'Duplicados de "binario": {duplicados_binario}')
`````

## Ejercicio 10
Convierte el programa del ejercicio 2 en una función al que se le pase el valor a buscar como un argumento posicional

`````python
def find_duplicates(lst, value_to_search):
    """
    Encuentra los índices en la lista `lst` donde aparece el `value_to_search`.
    
    Args:
    lst (list): La lista en la que buscar el valor.
    value_to_search (str): El valor cuya ubicación se busca.
    
    Returns:
    list: Una lista con los índices donde aparece el `value_to_search`.
    """
    indices = []
    value_to_search_lower = value_to_search.lower()  # Convertir el valor a minúsculas para comparación insensible a mayúsculas
    for index, value in enumerate(lst):
        if value.lower() == value_to_search_lower:  # Comparar el valor actual (en minúsculas) con el valor a buscar (en minúsculas)
            indices.append(index)  # Añadir el índice a la lista si hay una coincidencia
    return indices

# Lista de ejemplo
ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

# Probar la función con diferentes valores
print(find_duplicates(ejer_3, "un"))      # Salida esperada: [0, 7]
print(find_duplicates(ejer_3, "es"))      # Salida esperada: [3, 13]
print(find_duplicates(ejer_3, "binario")) # Salida esperada: [2, 15]
`````

## Ejercicio 12
Solicitar al usuario que ingrese su dirección email. Imprimir un mensaje indicando si la dirección es válida o no, valiéndose de una función para decidirlo. Una dirección se considerará válida si contiene el símbolo "@".

`````python
def es_email_valido(email):
    """
    Verifica si la dirección de correo electrónico es válida.
    
    Una dirección se considera válida si contiene el símbolo "@".
    
    Args:
    email (str): La dirección de correo electrónico a validar.
    
    Returns:
    bool: True si la dirección es válida, False en caso contrario.
    """
    return "@" in email

def main():
    # Solicitar al usuario que ingrese su dirección de email
    email_usuario = input("Ingrese su dirección de email: ")
    
    # Validar la dirección de email
    if es_email_valido(email_usuario):
        print("La dirección de email es válida.")
    else:
        print("La dirección de email no es válida.")

# Ejecutar la función principal
if __name__ == "__main__":
    main()
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def es_dni_valido(dni_completo):
    """
    Verifica si un DNI es válido.
    
    Un DNI se considera válido si cumple con las siguientes condiciones:
    1. Tiene entre 7 y 8 dígitos numéricos.
    2. La letra del DNI es correcta según el cálculo proporcionado.
    
    Args:
    dni_completo (str): El DNI completo en formato '12345678A' o '1234567A'.
    
    Returns:
    bool: True si el DNI es válido, False en caso contrario.
    """
    if len(dni_completo) < 8 or len(dni_completo) > 9:
        return False
    
    # Separar los números de la letra
    if len(dni_completo) == 9:
        numeros = dni_completo[:8]
        letra = dni_completo[8].upper()
    else:
        numeros = dni_completo[:7]
        letra = dni_completo[7].upper()
    
    # Verificar que todos los caracteres en los números sean dígitos
    if not numeros.isdigit():
        return False
    
    # Convertir la parte numérica a entero
    dni_numero = int(numeros)
    
    # Definir la cadena de letras
    palabra = 'TRWAGMYFPDXBNJZSQVHLCKE'
    
    # Calcular la letra esperada
    letra_correcta = palabra[dni_numero % 23]
    
    # Comparar la letra proporcionada con la letra calculada
    return letra == letra_correcta

# Probar la función con algunos ejemplos
print(es_dni_valido("55555555T"))  # Ejemplo válido
print(es_dni_valido("1234567Z"))   # Ejemplo válido
print(es_dni_valido("12345678A"))  # Ejemplo inválido
print(es_dni_valido("12345678"))   # Ejemplo inválido (falta la letra)
print(es_dni_valido("123456789"))  # Ejemplo inválido (demasiados dígitos)
`````
